## UNData API Exercise

In this exercise, you'll redo the data gathering phase of the UNData Exploration project by using APIs instead of downloading csv files.

You'll make use of the [World Bank Indicators API](https://datahelpdesk.worldbank.org/knowledgebase/articles/889392-about-the-indicators-api-documentation). Note that this API does not require an API key. Before attempting the exercise, it would be a good idea to skim through the Documentation page and to check out the [Basic Call Structure article](https://datahelpdesk.worldbank.org/knowledgebase/articles/898581).


```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
```

1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.


```python
endpointgdp = 'http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'
params = {'per_page': 17024}
response = requests.get(endpointgdp, params = params)
```

```python
gdpres = response.json()
```

```python
gdpres
```

```python
gdpres = pd.json_normalize(gdpres[1])
```

```python
gdpres
```

2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.


```python
endpointle = 'http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json'
params = {'per_page': 17024}
response = requests.get(endpointle, params=params)
```

```python
leres = response.json()
```

```python
leres
```

```python
leres = pd.json_normalize(leres[1])
```

```python
leres
```

Drop empty columns:

```python
gdpres = gdpres.drop(columns = ['unit', 'obs_status'], axis =1, errors='ignore')
```

```python
gdpres
```

```python
leres = leres.drop(columns = ['unit', 'obs_status'], axis =1, errors='ignore')
```

```python
leres
```

```python
gdpres.rename(columns={'countryiso3code':'country_code', 'value':'GDP_Per_Capita', 'indicator.id': 'indicator_id', 'indicator.value':'indicator_value', 'country.id':'country_id', 'country.value':'country_value'}, inplace=True)
```

```python
gdpres
```

```python
leres.rename(columns={'countryiso3code':'country_code', 'value':'Life_Expectancy', 'indicator.id': 'indicator_id', 'indicator.value':'indicator_value', 'country.id':'country_id', 'country.value':'country_value'}, inplace=True)
```

```python
leres
```

Merge resulting dataframes:

```python
gdp_le = pd.merge(gdpres, leres, how = 'outer')
```

```python
gdp_le
```

4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.


```python
endpointcountries = 'https://api.worldbank.org/v2/country/all?format=json'
params = {'per_page': 296}
response = requests.get(endpointcountries, params = params)
```

```python
countres = response.json()
```

```python
countres
```

```python
countres_df = pd.json_normalize(countres[1])
```

```python
countres_df
```

```python
gdp_le
```

```python
countres_df.rename(columns={'id':'country_code', 'iso2Code':'country_id'}, inplace=True)
```

```python
count_gdp_le_df = pd.merge(
    left = countres_df, 
    right = gdp_le,
    how = 'outer'
)
```
